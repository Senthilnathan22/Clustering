```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
X=dataset.iloc[:, [3,4]].values
```

```python
X.shape
```

```python
from sklearn.cluster import MeanShift
from sklearn.datasets import make_blobs
X_labels = make_blobs(n_samples=200, centers=5, random_state=42)
mean_shift = MeanShift()
mean_shift.fit(X)
plt.scatter(X[:, 0], X[:, 1], c=mean_shift.labels_, cmap='viridis', marker='o')
plt.scatter(mean_shift.cluster_centers_[:, 0], mean_shift.cluster_centers_[:, 1], 
            c='red', marker='x', s=200, linewidths=3, label='Cluster Centers')
plt.title('Mean Shift Clustering')
plt.xlabel('salary')
plt.ylabel('Spending')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import MeanShift
meanshift=MeanShift(bandwidth=5, seeds=None, bin_seeding=False,)
y_mean=meanshift.fit_predict(X)
```

```python
supervised=pd.DataFrame(dataset)
```

```python
supervised['Cluster_group']=y_mean
```

```python
supervised
```

```python
import seaborn as sns
facet=sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5],
                 fit_reg=False,legend=True, legend_out=True)
```
